```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 6, 4.5
```

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f)
plt.show()
```

```python
plt.plot(t, f)
plt.title('График f(t)')
plt.show()
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
plt.plot(t, f)
plt.title('График f(t)')
plt.show()
```

```python
title_dict = {'fontsize': 16, 'fontweight': 'bold', 'color': 'Blue', 'family': 'DejaVu Sans'}
```

```python
plt.plot(t, f)
plt.title('График f(t)', fontdict=title_dict)
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, color='green')
plt.title('График f(t)', fontdict=title_dict)
plt.xlabel('t')
plt.ylabel('f')
plt.show()
```
